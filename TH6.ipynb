```python
#W6A4
lst = list(input().split())
d = {}
for i in range(0, len(lst), 2):
    key = lst[i]
    value = lst[i+1]
    if key in d:
        d[key].append(value)
    else:
        d[key] = [value]
print(d)
```

```python
d1 ={}
s1 = int(input()) 
for i in range (s1):
    key = input()
    value = int(input())
    d1[key] = value

d2 = {}
s2 = int(input())
for i in range(s2):
    key = input()
    value = int(input())
    d2[key] = value

d3 = {}
for i in d1:
    d3[i] = d1[i]
for i in d2:
    if i in d3:
        d3[i] += d2[i]
    else:
        d3[i] = d2[i]
d4 = {}
for i in sorted(d3):
    d4[i] = d3[i]

print(d4)
```

```python
#W6A10
lst = list(map(int, input().split()))
k = int(input())
res = set()

for i in range (0, len(lst)):
    for j in range (i+1, len(lst)):
        if lst[i] + lst[j] == k:
            res.add(tuple(sorted((lst[i], lst[j]))))
res = list(res)
                    
print(sorted(res))
```

```python
#W6A12
lst = list(map(int, input().split()))
lstcnt = []
temp = 0
res = 0

for i in range (len(lst)):
    count = 1
    if lst[i] in lstcnt:
        continue
    lstcnt.append(lst[i])
    for j in range (i+1, len(lst)):
        if lst[i] == lst[j]:
            count += 1
    if count == temp and lst[i] < res:
         res = lst[i]    
    if count > temp:
            res = lst[i]
            temp = count
    
print(res)
```
