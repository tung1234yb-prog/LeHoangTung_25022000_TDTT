```python
#W6A18
n = int(input())
m = int(input())
k = int(input())

arr = []
for i in range (n):
    row = []
    for j in range (m):
        a = int(input())
        row.append(a)
    arr.append(row)
sum = 0
for i in arr:
    sum += i[k-1]

print(sum)
```

```python
#W6A19
def check(lst):
    n = lst[0]
    for i in range (len(lst)):
        if lst[i] > n:
            id = i
            n = lst[i]
    return id
 
lst = [1,2,3,8,4,5,6,7]
check(lst)
```

```python
#W6A20
def check(lst, k):
    for i in range (len(lst)):
        if lst[i] == k:
            return i
    else:
        return -1

lst = [1,2,3,4,5,6]
k = 7
check(lst, k)
```

```python
#W6A21
lst = list(map(str, input().split()))
dct = {}

for i in range (len(lst)):
    dct[lst[i]] = dct.get(lst[i], []) + [i]
for key in dct:
    dct[key] = tuple(dct[key])

print(dct)
```

```python
#W6A23
def remove(lst, x):
    newlst = []
    for i in lst:
        if i > x:
            newlst.append(i)
    return newlst

lst = [1,2,3,4]
x = 3
print(remove(lst, x))
 
        
```

```python
#W6A24
def abc(lst):
    lst1 = []
    lst2 = []
    for i in lst:
        if i != 0:
            lst1.append(i)
        else:
            lst2.append(i)
    return lst2 + lst1

lst = list(map(int,input().split()))
print(abc(lst))
```

```python
#W6A26
lst = list(map(int, input().split()))
sum = 0

for i in range (1, len(lst)-1):
    if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
        sum += lst[i]

print(sum)
```
