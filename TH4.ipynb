```python
#W4A6
try:    
    a, b = map(int, input('Nhap hai so nguyen duong').split())
    if a < b:
        flag = True
        tong = 0
        for i in range (a, b+1):
            if i < 2:
                continue
            for j in range (2, int(i**0.5) + 1):
                if i % j == 0:
                    flag = False
                    break
            if flag:
                tong += i
            flag = True 
        print('Tong cac so nguyen to trong doan [a, b] la', tong)
    else:
        print(f'So nguyen {a} phai nho hon so nguyen {b}') 
except ValueError:
    print(f'Ban phai nhap hai so nguyen')               
                
```

```python
#W4A7
try:
    n = int(input("Nhap mot so nguyen:"))
    uoc = 0
    if n >= 2:
        for i in range (n, 1, -1):
            flag = True
            for j in range (2, int(i**0.5) + 1):
                if i % j == 0:
                    flag = False
                    break
            if flag and n % i == 0:
                uoc = i
                break
        if uoc == 0:
            print(n, 'khong co uoc nguyen nao')
        else:
            print(n, 'co uoc nguyen lon nhat la', uoc)
    if n < 2:
        print('Ban phai nhap so nguyen lon hon 2')
except ValueError:
    print('Ban phai nhap mot so nguyen')
```

```python
#W4A8
n = int(input('Nhap mot so nguyen:'))
count = 0
while True:
    temp = n
    sodao = 0
    while temp > 0:
        sodao = sodao*10 + temp%10
        temp //= 10
    if n == sodao: 
        break
    n += sodao
    count += 1
print(f'So buoc la {count} va gia tri palindrome thu duoc la {n}')
```

```python
#W4A10
n = int(input('Nhap so nguyen duong n:'))
dodai = 0
x = 0
for i in range (1, n + 1):
    count = 1
    temp = i
    while temp != 1:
        if temp % 2 == 0:
           temp /= 2
        else: 
            temp = 3*temp + 1
        count += 1
    if count > dodai:
        dodai = count
        x = i
print(f'So co do dai day Collatz lon nhat la {x} va do dai cua day la {dodai}')
```
