```python
#Bai4
n = int(input())
for i in range (1, n):
  for j in range (1, n-i+1):
    print(" ", end = '')
  print('*', end = '')
  for k in range (1, 2*i-2):
    print(' ', end = '')
  if i > 1: 
    print('*')
  else:
    print('')
for i in range (1, 2*n):
   print('*', end = '')
```

```python
#Bai8
while True:
    a = int(input())
    b = int(input())
    if a > 0 and b > 0:
        break
if a < b:
    temp = a
    a = b
    b = temp
for i in range (b, 0, -1):
    if a % i == 0 and b % i == 0:
        print(i)
        break
    
```
