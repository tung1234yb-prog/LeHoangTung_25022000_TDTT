```python
#W5A1
def Max_Of_Two(a, b):
    if a > b:
        return a
    else:
        return b
```

```python
#W5A2
def swap(a, b):
    temp = a
    a = b
    b = temp
    return a, b
```

```python
#W5A3
def check(n):
    if n == 1:
        print(n, 'khong phai la so nguyen to')
    flag = True
    for i in range (2, int(n**(1/2))+1):
        if n % i == 0:
            flag = False
            break
    if flag:
        print(n, 'la so nguyen to')
    else: 
        print(n, 'khong phai la so nguyen to')
check(3)
    
```

```python
#W5A4
def check(n):
    tong = 0
    for i in range (1, int(n**(1/2))):
        if n % i == 0:
            tong += i
            tong += n//i
    return tong == n
check(100)
```

```python
#W5A5
def abc():
    lst = list(map(int, input().split()))
    while True:
        k = int(input())
        if k > 0:
            break
    for i in range (0, len(lst)):
        if lst[i] == k:
            return i
        else:
            return -1
```

```python
#W5A6
def gt(n):
    tich = 1
    for i in range (1, n+1):
        tich *= i
    return tich
```

```python
#W5A7
def cal(n1, n2, op):
    match op:
        case '+':
            res = n1 + n2
        case '-':
            res = n1 - n2
        case '*':
            res = n1 * n2
        case '/':
            res = n1 / n2
    return f'{res:.2f}'
```

```python
#W5A9
def sum(n):
    while n > 0:
        res = 0
        res += n % 10 
        n //= 10
    return res
```
